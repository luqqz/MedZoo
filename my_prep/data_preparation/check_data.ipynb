```python
import os
import numpy
import shutil
import skimage.transform
import SimpleITK

_final_data_processed_path = 'data_final_processed/'
NIFTI_extension = '.nii'

for root, _, files in os.walk(_final_data_processed_path, topdown=False):
    for file in files:
        if NIFTI_extension in file and 'ct' in file and 'train' in root:
            print("Processing: %s" % file)
            file_path = os.path.join(root, file)
            mask_path = file_path.replace('ct', 'mask')

            ct = SimpleITK.ReadImage(file_path)
            ct_size = numpy.array(ct.GetSize())[[2, 1, 0]]
            mask = SimpleITK.ReadImage(mask_path)
            mask_size = numpy.array(mask.GetSize())[[2, 1, 0]]
            
            if ct_size[0] != mask_size[0] or ct_size[1] != mask_size[1] or ct_size[2] != mask_size[2]:
                print("ERROR! ct_size = " + str(ct_size) + ", mask_size = " + str(mask_size) + "\nfile: " + file_path)
```
